# 4. Visualisation


#### 4.A Plot a histogram of `age`

````python
# A:
num_bins = 50

n, bins, patches = plt.hist(df['age'], bins=num_bins, facecolor='red', alpha=0.6)

plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
````

#### 4.B BoxPlot of `age`

````python
# A:

plt.boxplot(df['age'],vert=False, patch_artist=True,
           boxprops=dict(facecolor='lightblue', color='black'),
           flierprops=dict(markerfacecolor='red', marker='o', markersize=8))  #if its vertical its not necessary vert, can be column ='age' or df['age']
plt.title('BoxPlot of Age')
plt.ylabel('Age')
plt.show()
````

#### 4.C Create a scatterplot of `age` and `length_of_service`

````python
# A:
plt.scatter(df['age'],df['length_of_service'], color='red')
plt.title('ScatterPlot of Age')
plt.xlabel('Age')
plt.ylabel('Length of Service')
plt.show()
````
